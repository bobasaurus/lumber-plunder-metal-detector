```python
%matplotlib widget
from wavefunctions import GenerateSinewavePoints, GenerateSinewavePeriods
import matplotlib.pyplot as plt
import numpy as np

sampleRate = 500000.0
frequency = 10000.0
amplitude = 1.5

numCycles = 3

fig = plt.figure()
plt.axhline(y = 0.0, color = 'g', linestyle = '--') 

for phase in range(0, 100, 10):
    signal = amplitude * np.sin(np.linspace(0, 1.0/frequency * numCycles, int(sampleRate/frequency*numCycles)) * 2*np.pi*frequency + np.deg2rad(phase))
    plt.plot(signal, label="phase: " + str(phase))

plt.legend(loc="upper right")
```
