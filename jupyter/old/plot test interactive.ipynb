```python
%matplotlib widget

import matplotlib.pyplot as plt
import math
import numpy as np
from ipywidgets import interact, FloatSlider, IntSlider, Layout

fig = plt.figure()

def PlotBlah(amplitude):
    x = np.arange(0, math.pi*2, 0.05)
    y = amplitude * np.sin(x)
    print(amplitude)
    plt.clf()
    plt.plot(x, y)
    plt.gca().set_ylim([-1.05, 1.05])
    fig.canvas.draw()

interact(PlotBlah, 
         amplitude=FloatSlider(min=-1, max=1, step=0.01, value=-1, layout=Layout(width='100%')),
        )
```
